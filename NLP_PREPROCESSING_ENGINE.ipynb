**TASK-2**


```python
import re
def preprocess_text(text):
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    
    # Remove numbers
    text = re.sub(r"\d+", "", text)
    
    # Fix repeated characters
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)
    
    # Remove special characters
    text = re.sub(r"[^a-z\s]", "", text)
    
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    
    # Remove short words
    words = text.split()
    words = [word for word in words if len(word) > 2]
    
    return words
```

**TASK-3**

```python
sentences = [
"I absolutely looooved this 😍😍",
"Worst product ever!!! 😡",
"Visit http://example.com now",
"This is sooo coool!!!",
"Buy 2 items for 500 rupees",
"LOL this was funny 😂",
"Meeting at 10:30 AM",
"Check www.test.com link",
"Amazinggg service!!!",
"Ok"
]

for sentence in sentences:
    
    
    tokens = preprocess_text(sentence)
    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", " ".join(tokens))
```

**TASK-4**

```python
def token_analytics(sentences):
    
    for sentence in sentences:
        
        tokens = preprocess_text(sentence)
        
        if len(tokens) == 0:
            avg_length = 0
        else:
            avg_length = sum(len(word) for word in tokens) / len(tokens)
        
        print("\nSentence:", sentence)
        print("Total Tokens:", len(tokens))
        print("Unique Tokens:", len(set(tokens)))
        print("Average Token Length:", round(avg_length, 2))

token_analytics(sentences)
```

**TASK-5**

```python
from collections import Counter

all_tokens = []

for sentence in sentences:
    tokens = preprocess_text(sentence)
    all_tokens.extend(tokens)

word_counts = Counter(all_tokens)

print("\nTop 10 Frequent Words:")
print(word_counts.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(word_counts.most_common()[-5:])
```

**TASK-6**

```python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        
        tokens = preprocess_text(text)
        
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


result = full_pipeline(sentences)

print("\nPipeline Output:")
print(result)
```

```python
def safe_preprocess(text):

    if not text or not isinstance(text, str):
        return []

    tokens = preprocess_text(text)

    if len(tokens) == 0:
        return ["no_valid_text"]

    return tokens

test_cases = [
"",
"😂😂😂",
"123456"
]

for case in test_cases:

    print("\nInput:", case)
    print("Output:", safe_preprocess(case))
```
